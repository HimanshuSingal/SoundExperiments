```python
segment = 0.5
total_time = 60
total_number = int(total_time/segment)
frames = int(segment * 44100)
freq = np.fft.fftfreq(frames)
for i in range(total_number):
    fft_signal = np.fft.fft(s1[i*frames:(i+1)*frames])
    powers = []
    for i in fft_signal:
        powers.append(np.absolute(i))

    print(np.sum(powers))
    plt.plot(freq,powers)
    plt.show()
    plt.clf()
```
